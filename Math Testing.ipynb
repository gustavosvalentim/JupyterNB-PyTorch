```python
import torch
import matplotlib.pyplot as plt
```

```python
tensor = torch.randn(4, 2)
nparray = tensor.numpy()
```

```python
print(nparray)
```

```python
x = [x[0] for x in nparray]
y = [x[1] for x in nparray]
```

```python
x.sort(key = lambda x: x)
y.sort(key = lambda y: y)
```

```python
plt.plot(x, y)
```

```python
changingx = x[1] - x[0]
changingy = y[1] - y[0]
m = changingy / changingx
print(m)
```

```python
changingx2 = x[1] - x[2]
changingy2 = y[1] - y[2]
m2 = changingy2 / changingx2
print(m)
```

```python
plt.plot([m, m2])
```
